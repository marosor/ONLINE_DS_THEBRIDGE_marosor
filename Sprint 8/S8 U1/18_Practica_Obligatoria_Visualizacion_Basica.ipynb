```python
# Filtrar datos
ciudades_mas_pobladas = df_California_cities.nlargest(5, "population_total")
print(ciudades_mas_pobladas[["city","population_total"]].to_string(index = False))
```

```python
def ciudades_con_esp(ciudad): # Hago función porque son muchas ciudades y puede que necesite corregir otras también
    prefijos = ["San", "Santa", "Las", "Los", "New"]
    for prefijo in prefijos:
        if ciudad.startswith(prefijo):
            resto_ciu = ciudad[len(prefijo):]
            if resto_ciu[0].isupper():
                ciudad_esp =  ciudad[:len(prefijo)] + " " + resto_ciu
                return ciudad_esp
            break
    return ciudad

print(ciudades_con_esp("Santa Ana"))  # Debería imprimir "Santa Ana"
# por qué aquí si que me lo cambia pero en el gráfico no??????
```

```python
ciudades_esp = [ciudades_con_esp(ciudad) for ciudad in ciudades_mas_pobladas["city"].to_list()]
poblaciones = (ciudades_mas_pobladas["population_total"]/1000).to_list()

plt.bar(ciudades_esp, poblaciones);

fig, ax = plt.subplots(figsize = (10,10))
bars = ax.bar(ciudades_esp, poblaciones,color = ["blue", "grey", "red", "olive", "brown"], alpha = 0.5)
ax.set_title("Ciudades más pobladas de California", ha = "center", fontsize = 10)
ax.set_xlabel("Ciudades")
ax.set_ylabel("Habitantes (miles)")
ax.set_ylim(0,5000);

for bar in bars: # todo esto se lo he tenido que ir reguntando paso a paso a chatgpt, pero igual no me sale...
    pob_tot = bar.get_height() * 1000 # Revierto lo que dividí antes x 1000 para volver a la población original
    ax.annotate(f"{int(pob_tot)}",
                xy = (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                xytext = (0, 5),
                ha = "center")   
               
fig
# Ya probaré de hacerlo mejor en otro momento... queda todo junto superpuesto abajo de la primera barra...
```
